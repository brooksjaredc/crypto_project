```python
num_selections = 5
random_portfolios = []
er = []
risk = []

def er_and_risk(p, num):
    p_er = 0
    p_risk = 0
    #print(num, len(p))
    for j in range(num):
        p_er += (1.0/num)*np.mean(df3[p[j]])
        p_risk += (1.0/num)**2*np.var(df3[p[j]])
        for k in range(num):
            if(j!=k):
                p_risk += (1.0/num)**2*np.var(df3[p[j]])*df3[p[j]].corr(df3[p[k]])
    p_risk = np.sqrt(p_risk)
    
    return p_er, p_risk

for i in range(100):
    p = random.sample(top_curr, num_selections)
    p_er = 0
    p_risk = 0

    p_er, p_risk = er_and_risk(p, num_selections)
    er.append(p_er)
    risk.append(p_risk)
    random_portfolios.append(p)
    print(p, p_er, p_risk)
```

```python
p0 = random_portfolios[0]

curr_additions = []
increase_in_er = []
decrease_in_risk = []
new_er = []
new_risk = []

for c in top_curr:
    if(c in p0):
        continue
    p_er, p_risk = er_and_risk(p0, num_selections)
    #print(p0, c)
    p_new = p0.copy()
    p_new.append(c)
    #print(p_new, len(p_new))
    p_er_new, p_risk_new = er_and_risk(p_new, num_selections+1)
    if(p_risk_new>p_risk or p_er_new<p_er):
        continue
    curr_additions.append(c)
    increase_in_er.append(p_er_new-p_er)
    decrease_in_risk.append(p_risk-p_risk_new)
    new_er.append(p_er_new)
    new_risk.append(p_risk_new)
    
additions = pd.DataFrame(
    {'Symbol': curr_additions,
     'increase_in_er': increase_in_er,
     'decrease_in_risk': decrease_in_risk,
     'new_er': new_er,
     'new_risk': new_risk
    })
    
additions.sort_values(by='increase_in_er', ascending=False, inplace=True)
additions.reset_index(drop=True, inplace=True)
additions
```

```python
num_selections = 5
random_portfolios = []
portfolio_weights = []
er = []
risk = []
num_of_cryptos = [3,4,5,6]

def random_weights(n):
    weights = []
    for i in range(n):
        if(i==n-1):
            weights.append(1-sum(weights))
        else:
            high = 1 - sum(weights)
            weights.append(np.random.uniform(high=high))
    return(weights)
            
random_weights(5)
    
def er_and_risk(p, num, weights, df):
    p_er = 0
    p_risk = 0
    #print(num, len(p))
    for j in range(num):
        p_er += (weights[j])*np.mean(df[p[j]])
        p_risk += (weights[j])**2*np.var(df[p[j]])
        for k in range(num):
            if(j!=k):
                p_risk += (weights[j])*(weights[k])*np.std(df[p[j]])*np.std(df[p[k]])*df[p[j]].corr(df[p[k]])
    p_risk = np.sqrt(p_risk)
    
    return p_er, p_risk

for i in range(10):
    num = random.choice(num_of_cryptos)
    p = random.sample(top_curr, num)
    weights = random_weights(num)
    #print(num, len(p))
    
    p_er = 0
    p_risk = 0

#     p_er, p_risk = er_and_risk(p, num, weights, df3)
#     er.append(p_er)
#     risk.append(p_risk)

    portfolio_weights.append(weights)
    random_portfolios.append(p)
    print(p, weights)
```

```python
random_portfolios = []
portfolio_weights = []
num_of_cryptos = [3,4,5,6]

for i in range(500):
    num = random.choice(num_of_cryptos)
    p = random.sample(top_curr, num)
    weights = random_weights(num)

    portfolio_weights.append(weights)
    random_portfolios.append(p)
    #print(p, weights)
    
#er_and_risk(random_portfolios[0], len(random_portfolios[0]), portfolio_weights[0], df2_train_period)
```

```python
initial_er = []
initial_risk = []
delta_er = []
delta_risk = []
correlation = []
addition_er = []
addition_performance = []
addition_std = []
addition_marketcap = []
delta_future_er = []
delta_future_risk = []
did_risk_decrease = []
simple_prediction = []
portfolio_id = []

for i, portfolio in enumerate(random_portfolios):
    ini_er, ini_risk = er_and_risk(portfolio, len(portfolio), portfolio_weights[i], df2_train_period)
    future_er, future_risk = er_and_risk(portfolio, len(portfolio), portfolio_weights[i], df2_test_period)
    for c in top_curr:
        if(c in portfolio):
            continue
        new_portfolio = portfolio.copy()
        new_portfolio.append(c)
        new_weights = []
        for w in portfolio_weights[i]:
            new_weights.append(w*(1/1.2))
        new_weights.append(1-sum(new_weights))
        num = len(new_portfolio)
        new_er, new_risk = er_and_risk(new_portfolio, num, new_weights, df2_train_period)
        future_new_er, future_new_risk = er_and_risk(new_portfolio, num, new_weights, df2_test_period)
        
        portfolio_id.append(i)
        
        initial_er.append(ini_er)
        initial_risk.append(ini_risk)
        delta_er.append(new_er-ini_er)
        delta_risk.append(new_risk-ini_risk)
        
        addition_er.append(np.mean(df2_train_period[c]))
        addition_std.append(np.std(df2_train_period[c]))
        
        portfolio_performance = pd.Series([])
        for k in range(len(portfolio_weights[0])):
            new = portfolio_weights[0][k]*df2_train_period[random_portfolios[0][k]]
            portfolio_performance = portfolio_performance.add(new,fill_value=0)
        
        corr = portfolio_performance.corr(df2_train_period[c])
        correlation.append(corr)
        
        mc = df_train_period[df_train_period['Symbol']==c]['Market Cap'].iloc[-1]
        addition_marketcap.append(mc)
        delta_future_er.append(future_new_er-future_er)
        delta_future_risk.append(future_new_risk-future_risk)
        if(new_risk < ini_risk and new_er > ini_er):
            simple_prediction.append(1)
        else:
            simple_prediction.append(0)
        if(future_new_risk < future_risk and future_new_er > future_er):
            did_risk_decrease.append(1)
        else:
            did_risk_decrease.append(0)        
```
